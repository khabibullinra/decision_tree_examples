Пробуем нарисовать дерево с использованием матплотлиба

```python

import matplotlib.pyplot as plt

from matplotlib.patches import Circle, Polygon, Rectangle, FancyArrowPatch, FancyArrow, ArrowStyle, RegularPolygon

from matplotlib.offsetbox import (AnnotationBbox, DrawingArea, OffsetImage,
                                  TextArea)
```

```python
fig, ax = plt.subplots(figsize=(10, 6), layout='constrained')
ax.set_xlim(-1, 3)
ax.set_ylim(-1, 3)
ax.axis('off')


plt.text(0.6, 0.7, "eggs", size=10, 
         ha="center", va="center",
         bbox=dict(boxstyle="round",
                   ec=(1., 0.5, 0.5),
                   fc=(1., 0.8, 0.8),
                   )
         )

plt.text(0, 0, "eggs", size=10, 
         ha="center", va="center",
         bbox=dict(boxstyle="round",
                   ec=(1., 0.5, 0.5),
                   fc=(1., 0.8, 0.8),
                   )
         )

plt.text(1, 3, "eggs", size=10, 
         ha="center", va="center",
         bbox=dict(boxstyle="round",
                   ec=(1., 0.5, 0.5),
                   fc=(1., 0.8, 0.8),
                   )
         )

# Annotate the 1st position with a text box ('Test 1')
offsetbox = TextArea("Test 1")


# Define a 1st position to annotate (display it with a marker)
xy = (0, 0)
ax.plot(xy[0], xy[1], "r+")

ab = AnnotationBbox(offsetbox, xy,
                    xybox=(-30, 40),
                    xycoords='data',
                    boxcoords="offset points",
                    arrowprops=dict(arrowstyle="->"),
                    bboxprops=dict(boxstyle="round"))
ax.add_artist(ab)




plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6), layout='constrained')
ax.set_xlim(-1, 5)
ax.set_ylim(-1, 3)
ax.axis('off')



r = Rectangle((-1, 1), 0.4, 0.4, ec="b", fc='r')
p = RegularPolygon((3, 0), 3, radius=0.3, orientation=3.14/2, ec="b", fc='g')

с = Circle((0, 0), 0.2, ec="b", fc='y')
d = Circle((2, 2), 0.2, ec="none")


ax.add_artist(r)
ax.add_artist(с)
ax.add_artist(d)
ax.add_artist(p)

a = FancyArrowPatch((0, 0), (3, 0), 
                    arrowstyle="Simple, tail_width=1, head_width=12, head_length=13",
                    patchA = с,
                    patchB = p)


ax.add_artist(a)

plt.show()
```
